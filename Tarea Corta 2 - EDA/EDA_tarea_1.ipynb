### Procedencia del dataset
Este dataset es de la plataforma dataworld. Los datos que están plasmados son recolectados directamente de https://www.transfermarkt.co.uk/ en la cual se lleva un registro de traspasos de jugadores, equipos, juegos y otros datos más, como los que se quiere analizar son los datos relaciones entre jugadores y competiciones entonces se ignoran los otros datos disponibles.

### Tamaño del dataset
El dataset inicialmente consta de 20638 filas y 34 columnas.

### Significado de los atributos
- Crimen: como bien su nombre lo dice es el tipo de crimen que se perpetro. Atributo categórico.
- Subcrimen: es una especificación mayor del crimen, que tipo de armas se usaron o como se realizó el crimen. Atributo categórico.
- Hora inicio y Hora final: estos atributos representan el rango de 3 horas en las cuales se realizó el crimen. Atributo numérico.
- Victima: tipo de victima a la que le sucedio el crimen o que se vio afectada directamente. Atributo categórico.
- Subvictima: al igual que subcrimen es una especificación mayor de la victima. Atributo categórico.
- Edad: edad del perpetrador(menor de edad, mayor de edad o adulto mayor) Atributo categórico.
- Genero: el genero del perpetrador. Atributo categórico.
- Nacionalidad: la nacionalidad del perpetrador. Atributo categórico.
- Provincia: la provincia donde se realizó el crimen. Atributo categórico. 
- Canton: el canton donde se realizó el crimen. Atributo categórico.
- Año, Mes y Día: año, mes y día donde se realizó el crimen. Atributo numérico.
- DiaDeLaSemana: el día de la semana en el cual se realizo el crimen, siendo 0 lunes y 6 domingo. Atributo numérico.

### Matríz de correlación
Este mapa de correlación nos muestra que la cantidad de jugadores foráneos y el porcentaje de jugadores foráneos están fuertemente correlacionados. También el año de nacimiento y el identificador del jugador sin embargo por la naturaleza de estos dos atributos esta correlación se puede ignorar.

### Visualización de gráficas

```python
var = 'competitions.name'
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="year_of_birth", data=df)
plt.xticks(rotation=90);
```

#### Esta gráfica ayuda a responder la pregunta de investigación 2
Sin importar la competición el año promedio en el que los jugadores contratados nacieron fue entre 1990 y 1995, por lo que la competición no tiene inferencia alguna en la edad de los jugadores contratados.

```python
var = 'competitions.name'
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="height_in_cm", data=df)
plt.xticks(rotation=90);
```
